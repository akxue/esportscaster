```python
def tryOCR(image):
    img = Image.open('data/' + image)
    # invert
    inverted = ImageOps.invert(img)
    inverted.save('processed/inv' + image[:-4] + '.png')
    print(pytesseract.image_to_string(inverted))
    # threshold and convert to binary
    blackwhite = inverted.convert('LA')
    blackwhite.save('processed/bw' + image[:-4] + '.png')
    # print result
#     print(pytesseract.image_to_string(blackwhite))
    
```

```python
for file in listOfFiles:
    tryOCR(file)
```

```python
import cv2
img = cv2.imread('data/c90-4fox.png')
grayscaled = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
retval, threshold = cv2.threshold(grayscaled, 10, 255, cv2.THRESH_BINARY)
th = cv2.adaptiveThreshold(grayscaled, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 115, 1)
cv2.imshow('original',img)
cv2.imshow('Adaptive threshold',th)
cv2.imshow('grayscaled', grayscaled)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
